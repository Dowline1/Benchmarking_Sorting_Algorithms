### 1 - Bubble Sort Algorithm
Code for function adapted from (GeeksforGeeks, n.d.) using the optimized version.

```python
def bubbleSort(arr): 
    n = len(arr) 
   
    # Traverse through all array elements 
    for i in range(n): 
        swapped = False
  
        # Last i elements are already 
        #  in place 
        for j in range(0, n-i-1): 
   
            # traverse the array from 0 to 
            # n-i-1. Swap if the element  
            # found is greater than the 
            # next element 
            if arr[j] > arr[j+1] : 
                arr[j], arr[j+1] = arr[j+1], arr[j] 
                swapped = True
  
        # IF no two elements were swapped 
        # by inner loop, then break 
        if swapped == False: 
            break
```

### 2 - Heap Sort Algorithm
Code for function adapted from (GeeksforGeeks, n.d.).

```python
def heapify(arr, n, i): 
    largest = i  # Initialize largest as root 
    l = 2 * i + 1     # left = 2*i + 1 
    r = 2 * i + 2     # right = 2*i + 2 
  
    # See if left child of root exists and is 
    # greater than root 
    if l < n and arr[i] < arr[l]: 
        largest = l 
  
    # See if right child of root exists and is 
    # greater than root 
    if r < n and arr[largest] < arr[r]: 
        largest = r 
  
    # Change root, if needed 
    if largest != i: 
        arr[i],arr[largest] = arr[largest],arr[i]  # swap 
  
        # Heapify the root. 
        heapify(arr, n, largest)
        
def heapSort(arr): 
    n = len(arr) 
  
    # Build a maxheap. 
    for i in range(n, -1, -1): 
        heapify(arr, n, i) 
  
    # One by one extract elements 
    for i in range(n-1, 0, -1): 
        arr[i], arr[0] = arr[0], arr[i]   # swap 
        heapify(arr, i, 0) 
```

### 3 - Bucket Sort Algorithm
Code for function adapted from (Sanfoundry, n.d.).

```python
def bucketSort(alist):
    largest = max(alist)
    length = len(alist)
    size = largest/length
 
    buckets = [[] for _ in range(length)]
    for i in range(length):
        j = int(alist[i]/size)
        if j != length:
            buckets[j].append(alist[i])
        else:
            buckets[length - 1].append(alist[i])
 
    for i in range(length):
        insertion_sort(buckets[i])
 
    result = []
    for i in range(length):
        result = result + buckets[i]
 
 
def insertion_sort(alist):
    for i in range(1, len(alist)):
        temp = alist[i]
        j = i - 1
        while (j >= 0 and temp < alist[j]):
            alist[j + 1] = alist[j]
            j = j - 1
        alist[j + 1] = temp
```

### 4 - Shell Sort Algorithm
Code for function adapted from (GeeksforGeeks, n.d.).

```python
def shellSort(arr): 
  
    # Start with a big gap, then reduce the gap 
    n = len(arr) 
    gap = n//2
  
    # Do a gapped insertion sort for this gap size. 
    # The first gap elements a[0..gap-1] are already in gapped  
    # order keep adding one more element until the entire array 
    # is gap sorted 
    while gap > 0: 
  
        for i in range(gap,n): 
  
            # add a[i] to the elements that have been gap sorted 
            # save a[i] in temp and make a hole at position i 
            temp = arr[i] 
  
            # shift earlier gap-sorted elements up until the correct 
            # location for a[i] is found 
            j = i 
            while  j >= gap and arr[j-gap] >temp: 
                arr[j] = arr[j-gap] 
                j -= gap 
  
            # put temp (the original a[i]) in its correct location 
            arr[j] = temp 
        gap //= 2
```

### 5 - Gnome Sort Algorithm
Code for function adapted from (GeeksforGeeks, n.d.).

```python
def gnomeSort( arr, n): 
    index = 0
    while index < n: 
        if index == 0: 
            index = index + 1
        if arr[index] >= arr[index - 1]: 
            index = index + 1
        else: 
            arr[index], arr[index-1] = arr[index-1], arr[index] 
            index = index - 1
  
    return arr
```

### Introduction
Sorting algorithms in Python are functions which contain instructions and take an array as input, then perform sets of operations as specified in the function on the array and ouput the sorted area as per the instructions. At a high level there are two types of sorting algorithms <b>Comparison Based</b> sorting algorithms and <b>Non-Comparison Based</b> sorting Algorithms  (Moore, et al., n.d.).

#### Comarison Based Sorting Algorithms
A comparison based sorting algorithm sorts an array by comparing each element of the array with the other elements of the array to determine if the element should be to the left or right of another element in the array. Although they are simpler to implement, comparison based sorting algorithms cannot be faster than $O(n log n)$ (Moore, et al., n.d.). These algorithms use <b>Compartor Functions</b> (>, <, =) in order to build the logic of the algorithm using the functions to build rules.

#### Non-Comparison Based Sorting Algorithms
As the name suggests Non-Comparison Based Sorting Algorithms do not make comparisons between the elements of the array passed into the function and so can be faster than $O(n log n)$ (Moore, et al., n.d.). These types of algorithms work by determining for each element in the array how many elements are less than that element and places that element to the right of all lesser elements. The element is placed into this particular slot striaght away and so there is no need to rearrange the list.

#### Algorithm Complexity/Performance
When determining the complexity of an algorithm we consider two things <b>Time Complexty</b> and <b>Space Complexity</b> determination of these two features of an algorithm are part of determining <b>Algorithm Performance</b> (Kulalvaimozhi, et al., 2015).

##### Time Complexity <img style="float: right;" src="Images\clocks.jpg">
With regard to Time Complexity there are three types to consider which are best also called Big Oh ($O$), average also called Big Theta ($Θ$) and worst also called Big Omega ($Ω$) cases.

<b>Worst Case:</b>In this case the upper limit on running time of an algorithm is calculated which generally speaking occurs when the maximum number of operations getting executed over all of the inputs of size n. This occurs for example in Bubble sort if the array is reverse sorted the maximum number of comparisons occurs (Jain, 2018).

<b>Average Case:</b>In order to calculate average case times all possible inputs and calculate the computing time for these inputs. Next the calculated values are summed and divided by the total number of inputs, the distribution of the cases must either be known or predicted for all of the datasets of size n (Jain, 2018).

<b>Best Case:</b>For calculating best case the lower limit on running time of the algorithm i.e. the case which causes the minimum number of operations for an input of size n. In the case of Bubble Sort this occurs in the scenario whereby the input is already in a sorted state (Jain, 2018).

### Bubble Sort <img style="float: right;" width="250" src="Images\bubble.jpg">
The Bubble Sort algorithmis is a stable comparison based sorting algorithm and can be traced back as far as 1956 in (Friend, 1956) when it was reffered to as "sorting by exchange" and not "Bubble Sort". It was named for how the algorithm works on the array, the largest number in the array bubbles up to the highest element position or the right most position in the array just as a pot boils water (Campbell, 2020).

As a sorting algorithm Bubble Sort quite simply passes over each element in the array comparing each element to the element next to it. If the larger of the two elements is on the left then the element are swapped, this comparison is completed for each element until the entire array has been sorted and the largest element is in the highest or right most position of the array. In this manner the largest number is moved to the right until it is in the highest position, the algorithm then moves on until it finds the second largest element and moves it to the second highest position and so on an so forth (Campbell, 2020).

In order to sort an array of length n based on the above information it will take n passes of the algorithm over the array. The algorithm can also be further optimised by keeping track of the number of times the algorithm has passed over the array. Knowing this we can determine how many elements from the last have already been sorted i.e. after passing over the array 5 times we know that the last 5 elements of the array have been sorted so the algorithm must only asses up to the n-5 element (Campbell, 2020).

#### Time/Space Complexity
The worst case space complexity of Bubble Sort is $O(1)$ as the amount of space required for the algorithm is constant throughout the execution of the algorithm as the sorting is completed in place meaning only one temporary variable is being utilised.

In terms of time complexity in its best case  i.e. in the unlikely scenario when the array is already in a sorted state the time complexity will be $Ω(n)$ as the algorithm will only pass through the elements of the array once to confirm that it is already in a sorted state. 

The worst case for bubble sort will be $O(n^2)$ meaning that the array is sorted but in descending order, which means that the algorithm will need to pass through the entire array and swap each of the elements as it traverses the array n number of times in order to place each element in it's correct order. ### Heap Sort <img style="float: right;" width="250" src="Images\tree.jpg">
Heap Sort is an unstable comparison based sorting algorithmn which sorts arrays with the use of binary trees, it was developed in 1964 by J. W. J. Williams (Williams, 1964). A <b>Complete Binary Tree</b> is a binary tree where all levels with the possibility of the last is filled with all nodes of the tree being as far to the left as possible (GeeksforGeeks, n.d.).

Next moving onto a <b>Binary Heap</b> which is a complete binary tree that is sorted i.e. the value of every parent node is greater than it's 1/2 child node(s) with the largest value being the <b>Max Heap</b> and the smallest being the <b>Min Heap</b>. 

Heap sort executes as follows (Diagram Below Execution Steps): 
1. The array is first represented as a binary tree. 
2. A Max Heap is built from the data inputted as we assume the data is not sorted.
3. The Largest element of the max heap is swapped with the end or right most element of the array and removed from the binary tree with the removed element considered sorted.
4. The right most element that was swapped with the largest element is now at the top of the tree.
5. Heapify is then called which swaps the element at the top of the tree with the largest element in the tree.
6. If the element that was swapped is smaller than it's child nodes it is swapped with a child node until it is in a position where it is smaller than it's parent node.
7. We are now back to a Max Heap and the process executes again until every element has been sorted in this way.

#### Time/Space Complexity
The worst case space complexity of Heap Sort is $O(1)$ as the amount of space required for the algorithm is constant throughout the execution of the algorithm as the sorting is completed in place and only temporary variables are used to swap around the variables during the build of the max heap and during heapification.

In terms of time complexity heap sort has is $Ω(n log n)$ in best, worst and average cases. 

As the height of a complete binary tree containing n elements is $log(n)$, in the worst case scenario when an element must be moved from the root to the leaf node requires multiples of $log(n)$ comparisons and swaps.

While building the max heaps the algorithm computes this for $n/2$ elements giving a worst case complexity of around $n log n$.

While heap sort is actually sorting the root element is exchanged with the last element and the heapify follows. For each of the elements undergoing this the worst time results in $logn$ and this is then repeated n times and so results in a time complexity of $n log n$ asall steps are executed in sequence the complexity is not multiplied (Jajodia, 2017).

The worst case space complexity of Bucket Sort is $O(n x k)$ as the amount of space required for the algorithm is the number of elements inputed (n) times k which is the number of bucket created to house the elements (GeeksforGeeks, n.d.).

In terms of time complexity bucket sort has $O(n+k)$ in best case in the scenario where the elements get distributed perfectly amongst the buckets resulting in a fully sorted list without the need for insertion sort to organise the buckets (Growing with the Web, 2016).

### Gnome Sort <img style="float: right;" width="250" src="Images\gnome.jpg">
Gnome Sort is a stable comparison based sorting algorithm developed in 2000 by Hamid Sarbazi-Aza initially called "Stupid Sort" and referenced in his college newsletter(Computing Science Glasgow, 2000). The algorithm was later renamed to Gnome Sort as it was noted that the algorithm had likeness to the thought of a Garden Gnome sorting his flower pots.

The algorithm starts by recognising that it is at index 0 in that thre is no element to the left of it (start of pot line going back to the Gnome), the algorithm moves forward one index and compares the elements to its left and right. If the elements are in the correct order the algorithm moves forward, if the elements are not in order the elements are swapped and the algorithm steps back one index. If there are no elements in front of the algorithm the sorting has been completed (end of pot line) (GeeksforGeeks, n.d.).

#### Time/Space Complexity
Gnome sort has a space complexity of $O(1)$ as the algorithm only requires a constant temporary variable in order to make the swaps between the two array elements as it moves through the entire array sorting it in place on the way.

In the case of space complexity Gnome Sort has a best case of $O(n)$ in its best case whereby the array is already sorted as the algorithm will compare each pair of elements as it passes through the array making no swaps along the way.

For it's worst case time complexity Gnome sort is $Ω(𝑛2)$ and occurs when the array is sorted in reverse order and so the algorithm will need to swap every single pair of elements as it moves forward but also swap them as it steps back in the array after making the swap also.

```python
df = pd.DataFrame(array_size_list)
df["Bubble Sort"] = pd.DataFrame(bubble_sort_times)
df["Heap Sort"] = pd.DataFrame(heap_sort_times)
df["Bucket Sort"] = pd.DataFrame(bucket_sort_times)
df["Shell Sort"] = pd.DataFrame(shell_sort_times)
df["Gnome Sort"] = pd.DataFrame(gnome_sort_times)
df.columns = ['Array Size', 'Bubble Sort', 'Heap Sort', 'Bucket Sort', 'Shell Sort', 'Gnome Sort']
```
